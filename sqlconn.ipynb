```python
#importing libraries
!pip install mysql-connector-python
import pandas as pd
import mysql.connector
import numpy as np
```

```python
# csv to dataframe
df_buses_1=pd.read_csv("./Output_Files/df_buses_1.csv")
df_buses_2=pd.read_csv("./Output_Files/df_buses_2.csv")
df_buses_3=pd.read_csv("./Output_Files/df_buses_3.csv")
df_buses_4=pd.read_csv("./Output_Files/df_buses_4.csv")
df_buses_5=pd.read_csv("./Output_Files/df_buses_R.csv")
df_buses_6=pd.read_csv("./Output_Files/df_buses_SB.csv")
df_buses_7=pd.read_csv("./Output_Files/df_buses_H.csv")
df_buses_8=pd.read_csv("./Output_Files/df_buses_AS.csv")
df_buses_9=pd.read_csv("./Output_Files/df_buses_9.csv")
df_buses_10=pd.read_csv("./Output_Files/df_buses_10.csv")

Final_df=pd.concat([df_buses_1,df_buses_2,df_buses_3,df_buses_4,df_buses_5,df_buses_6,
                    df_buses_7,df_buses_8,df_buses_9,df_buses_10],ignore_index=True)
Final_df
```

# Data cleaning

```python
# data about the data
Final_df.info()
```

```python
#convert prices to numeric
Final_df["Price"]=Final_df["Price"].str.replace("INR","")
Final_df["Price"]=Final_df["Price"].astype(float)
Final_df["Price"].fillna(0) 
```

```python
#convert Ratings to numeric
Final_df["Ratings"]=Final_df["Ratings"].str.replace("New","")
Final_df["Ratings"]=Final_df["Ratings"].str.strip()
Final_df["Ratings"]=Final_df["Ratings"].str.split().str[0]
Final_df["Ratings"] = pd.to_numeric(Final_df["Ratings"], errors='coerce')
Final_df["Ratings"].fillna(0,inplace=True) 
```

```python
# info after the data type change
Final_df.info()
```

```python
Final_df = Final_df[Final_df["Price"] <= 7000]
```

```python
# replacing the nan value
Final_df = Final_df.replace({np.nan: None})
```

```python
# change dataframe to csv
path=r"./Output_Files/Final_busdetails_df.csv"
Final_df.to_csv(path,index=False)
```
